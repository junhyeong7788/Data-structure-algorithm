# 스택(Stack)
- 가장 마지막에 들어온 원소가 가장 먼저 추출되는 자료구조

```python
# 연결리스트를 이용한 구현

class Node:
    def __init__(self, data):
        self.data = data
        self.next = None


class Stack:
    def __init__(self):
        self.head = None

    # 원소 삽입
    def push(self, data):
        node = Node(data)
        node.next = self.head
        self.head = node

    # 원소 추출하기
    def pop(self):
        if self.is_empty():
            return None

        # 머리(head) 위치에서 노드 꺼내기
        data = self.head.data
        self.head = self.head.next

        return data

    # 최상위 원소(top)
    def top(self):
        if self.is_empty():
            return None
        return self.head.data

    # 먼저 추출할 원소부터 출력
    def show(self):
        cur = self.head
        while cur:
            print(cur.data, end=" ")
            cur = cur.next

    # 스택이 비어있는지 확인
    def is_empty(self):
        return self.head is None

stack = Stack()
arr = [1, 2, 3, 4, 5, 6, 7]
for x in arr:
    stack.push(x)
stack.show()
print()
```

```python
# 스택이 비어 있지 않을 때까지 반복하는 루프
while not stack.is_empty(): #not stack.is_empty() 조건이 참인 경우 코드블록을 계속 실행
    print(stack.pop()) # 스택의 최상위 요소를 출력하고 제거
```
